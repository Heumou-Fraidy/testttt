```python
documents = [

"""
Guide de gestion des déchets urbains à Yaoundé

La gestion des déchets urbains à Yaoundé constitue un enjeu majeur de santé publique et d’assainissement.
Les citoyens sont encouragés à participer activement à la propreté de leur environnement en signalant rapidement
les déchets non collectés et les dépôts sauvages.

Les principaux quartiers où des problèmes récurrents de collecte ont été observés incluent Mvog-Mbi, Bastos,
Nkolbisson, Emana et certains marchés fortement fréquentés.

Pour signaler un problème de collecte, les citoyens doivent fournir :
- la localisation exacte du problème ;
- le type de déchets observés ;
- la durée approximative du problème ;
- une photo si possible.

Les signalements sont transmis aux équipes de collecte compétentes afin de permettre une intervention rapide.
Les déchets ménagers doivent être déposés aux points de collecte prévus afin d’éviter l’accumulation anarchique.
""",

"""
Risques sanitaires liés aux déchets non collectés

Les déchets non collectés représentent un risque important pour la santé publique.
Lorsqu’ils restent longtemps dans les quartiers, ils peuvent favoriser le développement
de maladies infectieuses et la prolifération de nuisibles.

Parmi les principaux risques sanitaires figurent :
- le choléra ;
- le paludisme ;
- les infections respiratoires ;
- les maladies diarrhéiques ;
- les infections cutanées.

Les déchets stagnants attirent également des moustiques, des rats et d’autres animaux
pouvant transmettre des maladies.

Les autorités sanitaires recommandent :
- d’éviter le contact direct avec les déchets ;
- de ne pas brûler les déchets à proximité des habitations ;
- de protéger les enfants des zones contaminées ;
- de signaler rapidement tout dépôt sauvage.

Une intervention rapide des services de collecte permet de réduire les risques sanitaires.
""",

"""
Procédure officielle de signalement des déchets à Yaoundé

Lorsqu’un citoyen constate un problème de déchets non collectés,
il doit effectuer un signalement détaillé.

La procédure recommandée comprend les étapes suivantes :

1. Identifier précisément le lieu concerné.
2. Décrire le problème observé.
3. Estimer depuis combien de temps les déchets sont présents.
4. Fournir une image du site lorsque cela est possible.
5. Soumettre le signalement aux services compétents.

En cas d’accumulation importante près d’une école, d’un hôpital
ou d’un marché, le signalement peut être considéré comme prioritaire.

Après réception du signalement, une équipe de collecte peut être mobilisée
pour vérifier la situation et procéder à l’enlèvement des déchets.
""",

"""
Guide citoyen de bonnes pratiques environnementales

Les citoyens de Yaoundé jouent un rôle important dans la gestion des déchets urbains.
Afin de préserver un environnement sain, plusieurs bonnes pratiques sont recommandées.

Il est conseillé :
- de jeter les déchets dans les zones prévues ;
- de séparer les déchets recyclables lorsque cela est possible ;
- d’éviter les dépôts sauvages ;
- de respecter les jours de collecte ;
- de sensibiliser les voisins à la propreté.

La participation citoyenne améliore l’efficacité globale du système
de gestion des déchets et contribue à réduire les risques sanitaires.
"""
]

print("Nombre de documents :", len(documents))
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

# Configuration du découpage
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=300,
    chunk_overlap=50
)

# Création des chunks
docs = text_splitter.create_documents(documents)

# Vérification
print("Nombre total de chunks :", len(docs))
```

```python
for i, doc in enumerate(docs[:5]):
    print(f"\n--- Chunk {i+1} ---")
    print(doc.page_content)
```

```python
from langchain_huggingface import HuggingFaceEmbeddings

# Modèle léger et stable pour Colab
embedding_model = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)

print("Modèle embeddings chargé avec succès ✅")
```

```python
from langchain_community.vectorstores import Chroma

# Création de la base vectorielle
vectorstore = Chroma.from_documents(
    documents=docs,
    embedding=embedding_model
)

print("Base vectorielle créée avec succès ✅")
```

```python
# Création du retriever
retriever = vectorstore.as_retriever(
    search_kwargs={"k": 3}
)

print("Retriever créé avec succès ✅")
```

```python
question = "Quels sont les risques des déchets non collectés ?"

retrieved_docs = retriever.invoke(question)

for i, doc in enumerate(retrieved_docs):
    print(f"\n--- Résultat {i+1} ---")
    print(doc.page_content)
```

```python
def clean_yaounde_rag(question):

    # 1. Recherche des documents pertinents
    retrieved_docs = retriever.invoke(question)

    # 2. Construire le contexte
    context = "\n\n".join(
        [doc.page_content for doc in retrieved_docs]
    )

    # 3. Vérification si aucun contexte trouvé
    if not context.strip():
        return "Je n’ai pas trouvé cette information dans la base documentaire."

    # 4. Génération de réponse simple
    response = f"""
Voici les informations trouvées dans la base documentaire :

{context}

Question utilisateur :
{question}

Réponse :
Selon les documents disponibles, les informations ci-dessus répondent à votre demande.
"""

    return response
```

```python
def clean_yaounde_rag_v2(question):

    # Recherche des documents pertinents
    retrieved_docs = retriever.invoke(question)

    # Construire le contexte
    context = "\n".join([
        doc.page_content for doc in retrieved_docs
    ])

    # Vérifier si aucun document pertinent
    if len(retrieved_docs) == 0:
        return "Je n’ai pas trouvé cette information dans la base documentaire."

    # Réponses simples basées sur mots-clés
    question_lower = question.lower()

    # Cas : risques sanitaires
    if "risque" in question_lower or "maladie" in question_lower:
        return f"""
Selon les documents disponibles, les déchets non collectés présentent plusieurs risques sanitaires.

Les principaux risques identifiés sont :
- le choléra ;
- le paludisme ;
- les infections respiratoires ;
- les maladies diarrhéiques ;
- certaines infections cutanées.

Les autorités recommandent d’éviter le contact direct avec les déchets
et de signaler rapidement les zones concernées.
"""

    # Cas : signalement
    elif "signaler" in question_lower or "signalement" in question_lower:
        return f"""
Selon la procédure officielle de gestion des déchets à Yaoundé :

Pour signaler un problème de déchets, le citoyen doit :
1. indiquer la localisation exacte ;
2. décrire le problème observé ;
3. préciser depuis combien de temps les déchets sont présents ;
4. fournir une photo si possible.

Le signalement est ensuite transmis aux équipes compétentes.
"""

    # Cas : bonnes pratiques
    elif "bonne pratique" in question_lower or "environnement" in question_lower:
        return f"""
Les bonnes pratiques recommandées sont :
- utiliser les points de collecte prévus ;
- éviter les dépôts sauvages ;
- respecter les jours de collecte ;
- sensibiliser les voisins à la propreté.
"""

    # Cas par défaut
    elif len(context.strip()) > 50:
        return f"""
Voici les informations trouvées dans la base documentaire :

{context}
"""

    else:
        return "Je n’ai pas trouvé cette information dans la base documentaire."
```

```python
def clean_yaounde_rag_v3(question):

    # Recherche avec score de similarité
    retrieved_docs = vectorstore.similarity_search_with_score(
        question,
        k=3
    )

    # Vérifier pertinence minimale
    relevant_docs = []

    for doc, score in retrieved_docs:
        # Plus le score est petit, plus c'est pertinent
        if score < 1.2:
            relevant_docs.append(doc)

    # Aucun document pertinent trouvé
    if len(relevant_docs) == 0:
        return "Je n’ai pas trouvé cette information dans la base documentaire."

    # Construire le contexte
    context = "\n".join([
        doc.page_content for doc in relevant_docs
    ])

    question_lower = question.lower()

    # Cas risques sanitaires
    if "risque" in question_lower or "maladie" in question_lower:
        return """
Selon les documents disponibles, les déchets non collectés peuvent provoquer :

- le choléra ;
- le paludisme ;
- des infections respiratoires ;
- des maladies diarrhéiques ;
- certaines infections cutanées.

Il est recommandé d’éviter le contact direct avec les déchets et de signaler rapidement les zones concernées.
"""

    # Cas signalement
    elif "signaler" in question_lower or "signalement" in question_lower:
        return """
Pour signaler un problème de déchets à Yaoundé, le citoyen doit :

1. préciser la localisation exacte ;
2. décrire le problème ;
3. indiquer la durée du problème ;
4. fournir une photo si possible.

Le signalement est ensuite transmis aux équipes compétentes.
"""

    # Cas bonnes pratiques
    elif "bonne pratique" in question_lower or "environnement" in question_lower:
        return """
Les bonnes pratiques recommandées incluent :

- utiliser les points de collecte ;
- éviter les dépôts sauvages ;
- respecter les jours de collecte ;
- sensibiliser les voisins.
"""

    # Cas général
    return f"""
Informations trouvées dans la base documentaire :

{context}
"""
```

```python
print(clean_yaounde_rag_v3(
    "Qui est le président du Brésil ?"
))
```

```python
def clean_yaounde_rag_v4(question):

    # Recherche avec score
    results = vectorstore.similarity_search_with_score(question, k=3)

    # Trier par pertinence
    results = sorted(results, key=lambda x: x[1])

    best_doc, best_score = results[0]

    # 🔥 SEUIL DE CONFIANCE PLUS STRICT
    # (plus le score est petit, plus c’est pertinent)
    if best_score > 0.85:
        return "Je n’ai pas trouvé cette information dans la base documentaire."

    # Construire contexte uniquement si pertinent
    context = "\n".join([doc.page_content for doc, score in results])

    q = question.lower()

    # Cas risques
    if "risque" in q or "maladie" in q:
        return """
Selon les documents disponibles, les déchets non collectés peuvent provoquer :
- choléra
- paludisme
- infections respiratoires
- maladies diarrhéiques
- infections cutanées

Il est recommandé de signaler rapidement les zones concernées.
"""

    # Cas signalement
    elif "signaler" in q:
        return """
Pour signaler des déchets à Yaoundé :
1. Localisation exacte
2. Description du problème
3. Durée du problème
4. Photo si possible

Le signalement est transmis aux équipes de collecte.
"""

    # Cas bonnes pratiques
    elif "pratique" in q:
        return """
Bonnes pratiques :
- utiliser les points de collecte
- éviter les dépôts sauvages
- respecter les jours de collecte
- sensibiliser les voisins
"""

    # Cas général contrôlé
    return f"""
Informations trouvées dans la base documentaire :

{best_doc.page_content}
"""
```
